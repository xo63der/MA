```python
import SimpleITK as sitk
import os

import brain_functions as bf

def biascorrection(input_path, output_path, val=200):
    listing = os.listdir(input_path)

    for file in listing:
        inputImage = sitk.ReadImage(input_path+file)

        maskImage = sitk.OtsuThreshold( inputImage, 0, 1, val )

        inputImage = sitk.Cast( inputImage, sitk.sitkFloat32 )
        corrector = sitk.N4BiasFieldCorrectionImageFilter();
        output = corrector.Execute( inputImage, maskImage )

        sitk.WriteImage(output, output_path+file)

        print(file+'done')

preprocess_train = True
preprocess_test = False

if preprocess_train:
    
    input_path = "/home/hoffmann/MRI/preprocessed/train_data/"
    output_path = "/home/hoffmann/MRI/preprocessed/pre_train_data/"

    biascorrection(input_path, output_path)
    
    insert=bf.load_whole(input_path)
    output=bf.load_whole(output_path)

    bf.show_whole(insert)
    bf.show_whole(output)
    
if preprocess_test:
    
    input_path = "/home/hoffmann/MRI/preprocessed/test_data/"
    output_path = "/home/hoffmann/MRI/preprocessed/pre_test_data/"

    biascorrection(input_path, output_path)
    
    insert=bf.load_whole(input_path)
    output=bf.load_whole(output_path)

    bf.show_whole(insert)
    bf.show_whole(output)

    
```
